```python
%matplotlib inline

# Author: Cory Weber (cweber@anl.gov)
# License: BSD 3 clause
import matplotlib.pyplot as plt
import pyart
import numpy as np
import scipy.ndimage as spyi

sweep = 0

# read file
filename = '/media/meso/DATA/phd/obs/radar/CP2/mdv/s_band/rhi/20141106/031202.mdv'
radar = pyart.io.read_mdv(filename)
display = pyart.graph.RadarDisplay(radar)
fig = plt.figure(figsize=[20, 5])
# plot velocity
ax = fig.add_subplot(1,1,1)
display.plot('velocity', sweep=sweep, vmin=-15, vmax=15, fig=fig, ax=ax, colorbar_label='Velocity (m/s)')
display.set_limits(xlim=[0, 50], ylim=[0, 5])
```
